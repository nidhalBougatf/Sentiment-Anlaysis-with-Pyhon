```python
import nltk
nltk.download('vader_lexicon')
def nltk_sentiment(sentence):
    from nltk.sentiment.vader import SentimentIntensityAnalyzer
    
    nltk_sentiment = SentimentIntensityAnalyzer()
    score = nltk_sentiment.polarity_scores(sentence)
    return score

```

```python
nltk_results = [nltk_sentiment(row) for row in df['text']]
results_df = pd.DataFrame(nltk_results)
text_df = pd.DataFrame(df, columns = ['text'])
nltk_df = text_df.join(results_df)
```

```python

#int(NewValue)
def modularity(list):
    result=[]
    OldRange = (1 - -1)  
    NewRange = (5 - -5)  
    NewValue = (((-0.5 - -1) * NewRange) / OldRange) + -5
    
    for index, row in list.iterrows():
        #if(row['Weight']>1 or row['Weight']<-1):
            #list.iloc[index,3]=1
        #display(row['Weight'])
        weigh=int((((row['Weight'] - -1) * NewRange) / OldRange) + -5)
        if(weigh==0 and row['Weight']>0):
            result.append(1)
        elif(weigh==0 and row['Weight']==0):
            result.append(1)
        elif (weigh==0 and row['Weight']<0):
            result.append(-1)
            #list2.iloc[i,2]=int((((row['feeling'] - -1) * NewRange) / OldRange) + -5)
        else:
            
            result.append(weigh)
        #list[index]=int((((list[index] - -1) * NewRange) / OldRange) + -5)
    return result
```
